```python
"""
token_cnn_bayes_metaoptnet.py

Few-shot AMR group classification with:
- Token-CNN k-mer encoder
- Bayesian MetaOptNet-style ridge regression head
- 3-way 3-shot 5-query episodic meta-training

This extends your MetaOptNet-Ridge by:
- putting a Gaussian prior on the classifier weights W
- computing a Bayesian posterior per episode
- yielding per-query predictive variance (uncertainty)
"""

# ============================================================
# 0) Imports, seeds, device
# ============================================================
import os, random, numpy as np, pandas as pd
import torch, torch.nn as nn, torch.nn.functional as F
from tqdm import tqdm
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
random.seed(RANDOM_SEED)
torch.manual_seed(RANDOM_SEED)
if torch.cuda.is_available():
    torch.cuda.manual_seed(RANDOM_SEED)
    torch.backends.cudnn.deterministic = True

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device:", device)

# ============================================================
# 1) Load & clean MEGARes processed CSV (GROUP-LEVEL)
# ============================================================
CSV_PATH = "megares_fasta_processed.csv"  # adjust if needed
LABEL_COL = "group"   # group-level hierarchy
MIN_COUNT = 10        # min sequences per group to keep

df = pd.read_csv(CSV_PATH)

df = df.dropna(subset=["sequence", LABEL_COL]).reset_index(drop=True)
df["sequence"] = df["sequence"].str.upper().str.replace(r"[^ACGT]", "", regex=True)

print("Raw unique groups:", df[LABEL_COL].nunique())

group_counts = df[LABEL_COL].value_counts()
keep_groups = group_counts[group_counts >= MIN_COUNT].index
df = df[df[LABEL_COL].isin(keep_groups)].reset_index(drop=True)

print("Filtered dataset:", df.shape)
print("Filtered unique groups:", df[LABEL_COL].nunique())

if "mechanism" in df.columns and "drug_class" in df.columns:
    print("Example group → mechanism / drug_class:")
    print(df[["group", "mechanism", "drug_class"]].drop_duplicates().head())

# ============================================================
# 2) Train/Val/Test split by GROUP (not mechanism)
# ============================================================
labels = np.array(sorted(df[LABEL_COL].unique()))
train_labels, temp_labels = train_test_split(
    labels, test_size=0.30, random_state=RANDOM_SEED, shuffle=True
)
val_labels, test_labels = train_test_split(
    temp_labels, test_size=0.50, random_state=RANDOM_SEED, shuffle=True
)

train_df = df[df[LABEL_COL].isin(train_labels)].reset_index(drop=True)
val_df   = df[df[LABEL_COL].isin(val_labels)].reset_index(drop=True)
test_df  = df[df[LABEL_COL].isin(test_labels)].reset_index(drop=True)

print(f"Train/Val/Test {LABEL_COL}s: {len(train_labels)}/{len(val_labels)}/{len(test_labels)}")
print(f"Train/Val/Test samples: {len(train_df)}/{len(val_df)}/{len(test_df)}")

# ============================================================
# 3) Tokenization (train-only vocab) - k-mers
# ============================================================
KMER_K = 5
PAD, UNK = "<PAD>", "<UNK>"

def kmers_from_seq(seq, k=KMER_K):
    L = len(seq)
    if L < k:
        return []
    return [seq[i:i+k] for i in range(L-k+1)]

from collections import Counter
counter = Counter()
for s in train_df["sequence"]:
    counter.update(kmers_from_seq(s, KMER_K))

tokens = [PAD, UNK] + sorted(counter.keys())
stoi = {t: i for i, t in enumerate(tokens)}
itos = {i: t for t, i in stoi.items()}
VOCAB_SIZE = len(stoi)
print("Vocab size:", VOCAB_SIZE)

def encode_ids(seq, k=KMER_K):
    arr = kmers_from_seq(seq, k)
    if not arr:
        return [stoi[UNK]]
    return [stoi.get(tok, stoi[UNK]) for tok in arr]

def encode_df_to_ids(dfp):
    ids = [encode_ids(s, KMER_K) for s in dfp["sequence"].tolist()]
    nums = dfp[["gc_content","seq_len"]].to_numpy(np.float32)
    y = dfp[LABEL_COL].to_numpy()
    return ids, nums, y

train_ids, train_num, ytr = encode_df_to_ids(train_df)
val_ids,   val_num,   yva = encode_df_to_ids(val_df)
test_ids,  test_num,  yte = encode_df_to_ids(test_df)

# ============================================================
# 4) Padding, RC tokens, numeric standardization
# ============================================================
def pick_max_len(list_of_ids, pct=0.90):
    lens = np.array([len(x) for x in list_of_ids])
    return int(np.clip(np.quantile(lens, pct), 16, 2048))

MAX_LEN = 512
print("MAX_LEN:", MAX_LEN)

def pad_sequences(list_of_ids, max_len=MAX_LEN, pad_id=None):
    if pad_id is None:
        pad_id = stoi[PAD]
    out = np.full((len(list_of_ids), max_len), pad_id, dtype=np.int64)
    for i, seq in enumerate(list_of_ids):
        s = seq[:max_len]
        out[i, :len(s)] = s
    return out

Xtr_tok = pad_sequences(train_ids, MAX_LEN)
Xva_tok = pad_sequences(val_ids,   MAX_LEN)
Xte_tok = pad_sequences(test_ids,  MAX_LEN)

_comp = str.maketrans("ACGT", "TGCA")
def rev_comp(seq):
    return seq.translate(_comp)[::-1]

train_ids_rc = [encode_ids(rev_comp(s), KMER_K) for s in train_df["sequence"].tolist()]
Xtr_tok_rc   = pad_sequences(train_ids_rc, MAX_LEN)

num_mean = train_num.mean(axis=0, keepdims=True)
num_std  = train_num.std(axis=0, keepdims=True) + 1e-6
train_num = (train_num - num_mean)/num_std
val_num   = (val_num   - num_mean)/num_std
test_num  = (test_num  - num_mean)/num_std

# ============================================================
# 5) Episodic sampler with RC augmentation
# ============================================================
from collections import defaultdict

def mech_index(y):
    d = defaultdict(list)
    for i, lab in enumerate(y):
        d[lab].append(i)
    return {k: np.asarray(v, dtype=int) for k, v in d.items()}

def _choose_tokens_with_rc(Xtok, Xtok_rc, ids, rng, rc_prob=0.25):
    if Xtok_rc is None or rc_prob <= 0:
        return Xtok[ids]
    mask = rng.random(len(ids)) < rc_prob
    out = Xtok[ids].copy()
    if mask.any():
        out[mask] = Xtok_rc[ids[mask]]
    return out

def create_tasks(
    X_tok, X_num, y, idx_map,
    num_tasks=1000, N=3, K=3, Q=5,
    seed=42, X_tok_rc=None, rc_prob=0.25, augment_rc=False
):
    rng = np.random.default_rng(seed)
    tasks = []
    valid = [m for m, ids in idx_map.items() if len(ids) >= K + Q]
    if len(valid) < N:
        return tasks

    for _ in range(num_tasks):
        me_sel = rng.choice(valid, size=N, replace=False)
        s_tok, s_num, s_y = [], [], []
        q_tok, q_num, q_y = [], [], []
        for j, m in enumerate(me_sel):
            ids = rng.choice(idx_map[m], size=K+Q, replace=False)
            s, q = ids[:K], ids[K:K+Q]

            if augment_rc and (X_tok_rc is not None):
                s_tok.append(_choose_tokens_with_rc(X_tok, X_tok_rc, s, rng, rc_prob))
                q_tok.append(_choose_tokens_with_rc(X_tok, X_tok_rc, q, rng, rc_prob))
            else:
                s_tok.append(X_tok[s])
                q_tok.append(X_tok[q])

            s_num.append(X_num[s])
            q_num.append(X_num[q])
            s_y.append(np.full(K, j, np.int64))
            q_y.append(np.full(Q, j, np.int64))

        tasks.append({
            "s_tok": np.vstack(s_tok),
            "s_num": np.vstack(s_num),
            "s_y":   np.concatenate(s_y),
            "q_tok": np.vstack(q_tok),
            "q_num": np.vstack(q_num),
            "q_y":   np.concatenate(q_y),
            "mechs": list(me_sel)
        })
    return tasks

idx_tr = mech_index(ytr)
idx_va = mech_index(yva)
idx_te = mech_index(yte)

N, K, Q = 3, 3, 5
RC_PROB = 0.25
SUPPORT_DROP = 0.03
QUERY_DROP   = 0.01

train_tasks = create_tasks(
    Xtr_tok, train_num, ytr, idx_tr,
    num_tasks=1500, N=N, K=K, Q=Q,
    seed=RANDOM_SEED, X_tok_rc=Xtr_tok_rc,
    rc_prob=RC_PROB, augment_rc=True
)
val_tasks = create_tasks(
    Xva_tok, val_num, yva, idx_va,
    num_tasks=300, N=N, K=K, Q=Q,
    seed=RANDOM_SEED, augment_rc=False
)
test_tasks = create_tasks(
    Xte_tok, test_num, yte, idx_te,
    num_tasks=500, N=N, K=K, Q=Q,
    seed=RANDOM_SEED, augment_rc=False
)

print(f"Tasks | train:{len(train_tasks)} val:{len(val_tasks)} test:{len(test_tasks)}  (N={N},K={K},Q={Q})")

# ============================================================
# 6) TokenDropout
# ============================================================
def token_dropout(arr, p=0.0, pad_id=0):
    if p <= 0:
        return arr
    mask = (np.random.rand(*arr.shape) < p)
    out = arr.copy()
    out[mask] = pad_id
    return out

# ============================================================
# 7) Token-CNN encoder
# ============================================================
class CNNSeqEncoder(nn.Module):
    def __init__(
        self, vocab_size, pad_idx, max_len,
        embed_dim=256, token_dim=128,
        conv_channels=96, kernel_sizes=(3, 5, 7),
        use_cosine=True, num_features=2
    ):
        super().__init__()
        self.use_cosine = use_cosine
        self.pad_idx = pad_idx

        self.emb = nn.Embedding(vocab_size, token_dim, padding_idx=pad_idx)
        self.emb_dropout = nn.Dropout(0.20)

        self.convs = nn.ModuleList([
            nn.Sequential(
                nn.Conv1d(token_dim, conv_channels, kernel_size=k, padding=k//2),
                nn.ReLU(),
                nn.Dropout1d(0.10)
            ) for k in kernel_sizes
        ])

        self.num_proj = nn.Sequential(
            nn.Linear(num_features, 32),
            nn.ReLU(),
            nn.BatchNorm1d(32),
        )

        fused_in = conv_channels * len(kernel_sizes) * 2 + 32
        self.proj = nn.Sequential(
            nn.Linear(fused_in, 512),
            nn.ReLU(),
            nn.BatchNorm1d(512),
            nn.Dropout(0.50),
            nn.Linear(512, embed_dim)
        )

        self.log_temp = nn.Parameter(torch.zeros(1))

    def forward(self, tokens, nums):
        x = self.emb(tokens)          # [B,L,E]
        x = self.emb_dropout(x)
        x = x.transpose(1, 2)         # [B,E,L]

        pooled = []
        for block in self.convs:
            h = block(x)                              # [B,C,L]
            p_max  = F.adaptive_max_pool1d(h, 1).squeeze(-1)
            p_mean = F.adaptive_avg_pool1d(h, 1).squeeze(-1)
            pooled.extend([p_max, p_mean])

        h_text = torch.cat(pooled, dim=1)
        h_num  = self.num_proj(nums)
        h      = torch.cat([h_text, h_num], dim=1)

        z = self.proj(h)
        if self.use_cosine:
            z = F.normalize(z, p=2, dim=1)
        return z

# ============================================================
# 8) Bayesian MetaOptNet ridge head
# ============================================================
class BayesianRidgeHead(nn.Module):
    """
    Bayesian linear classifier per episode:
    W ~ N(0, tau^2 I),  y | W ~ N(ZW, sigma^2 I)

    We compute:
      A = (1/sigma^2) Z^T Z + (1/tau^2) I
      A_inv = A^{-1}
      mu_W  = A_inv (1/sigma^2) Z^T Y

    For a query z*:
      mean logits  = z*^T mu_W
      var per z*   = sigma^2 + z*^T A_inv z*
    """
    def __init__(self, init_log_tau2=-2.0, init_log_sigma2=0.0):
        super().__init__()
        self.log_tau2   = nn.Parameter(torch.tensor(init_log_tau2))   # prior variance
        self.log_sigma2 = nn.Parameter(torch.tensor(init_log_sigma2)) # noise variance

    def forward(self, s_z, s_y, q_z):
        """
        s_z: [S,d], s_y: [S], q_z: [Q,d]
        Returns:
            logits_mean: [Q,C]
            var_per_query: [Q]  (scalar variance per query, shared across classes)
        """
        device = s_z.device
        S, d = s_z.shape
        num_classes = int(s_y.max().item()) + 1

        Y = F.one_hot(s_y, num_classes=num_classes).float()  # [S,C]
        Z = s_z                                              # [S,d]

        tau2   = self.log_tau2.exp()
        sigma2 = self.log_sigma2.exp()

        A = (Z.t() @ Z) / sigma2 + torch.eye(d, device=device) / tau2    # [d,d]
        A_inv = torch.linalg.inv(A)                                      # [d,d]

        B = (Z.t() @ Y) / sigma2                                         # [d,C]
        W_mean = A_inv @ B                                               # [d,C]

        logits_mean = q_z @ W_mean                                       # [Q,C]

        qA = q_z @ A_inv                                                 # [Q,d]
        var_per_query = sigma2 + (qA * q_z).sum(dim=1)                   # [Q]

        return logits_mean, var_per_query

# ============================================================
# 9) Evaluation helper
# ============================================================
@torch.no_grad()
def evaluate_bayes_metaoptnet(encoder, head, tasks, device):
    encoder.eval(); head.eval()
    accs, losses = [], []
    for t in tasks:
        s_tok = torch.from_numpy(t["s_tok"]).long().to(device)
        s_num = torch.from_numpy(t["s_num"]).float().to(device)
        q_tok = torch.from_numpy(t["q_tok"]).long().to(device)
        q_num = torch.from_numpy(t["q_num"]).float().to(device)
        s_y   = torch.from_numpy(t["s_y"]).long().to(device)
        q_y   = torch.from_numpy(t["q_y"]).long().to(device)

        s_z = encoder(s_tok, s_num)
        q_z = encoder(q_tok, q_num)

        logits_q, q_var = head(s_z, s_y, q_z)
        loss = F.cross_entropy(logits_q, q_y)

        preds = logits_q.argmax(dim=1)
        accs.append((preds == q_y).float().mean().item())
        losses.append(loss.item())

    return float(np.mean(accs)), float(np.std(accs)), float(np.mean(losses))

# ============================================================
# 10) Meta-training: Token-CNN + Bayesian MetaOptNet-Ridge
# ============================================================
EMBED_DIM = 256
encoder = CNNSeqEncoder(
    vocab_size=VOCAB_SIZE,
    pad_idx=stoi[PAD],
    max_len=MAX_LEN,
    embed_dim=EMBED_DIM,
    token_dim=128,
    conv_channels=96,
    kernel_sizes=(3, 5, 7),
    use_cosine=True,
    num_features=2
).to(device)

bayes_head = BayesianRidgeHead().to(device)

LR = 2e-4
opt   = torch.optim.AdamW(
    list(encoder.parameters()) + list(bayes_head.parameters()),
    lr=LR, weight_decay=2e-4
)
sched = torch.optim.lr_scheduler.CosineAnnealingLR(opt, T_max=800)

EPISODES   = 1400
EVAL_EVERY = 20
PATIENCE   = 20

print({
    "model": "Token-CNN + Bayesian MetaOptNet-Ridge",
    "cosine_norm": encoder.use_cosine,
    "rc_train": True,
    "rc_prob": RC_PROB,
    "max_len": MAX_LEN,
    "k": KMER_K,
    "dropout_support": SUPPORT_DROP,
    "dropout_query": QUERY_DROP,
    "lr": LR,
    "episodes": EPISODES,
})

best_val, bad = 0.0, 0
train_losses, train_accs, val_accs = [], [], []
PAD_ID = stoi[PAD]

print("Training Bayesian MetaOptNet-Ridge...")
for ep in tqdm(range(1, EPISODES + 1)):
    encoder.train(); bayes_head.train()
    t = random.choice(train_tasks)

    t_s_tok = token_dropout(t["s_tok"], p=SUPPORT_DROP, pad_id=PAD_ID)
    t_q_tok = token_dropout(t["q_tok"], p=QUERY_DROP,   pad_id=PAD_ID)

    s_tok = torch.from_numpy(t_s_tok).long().to(device)
    s_num = torch.from_numpy(t["s_num"]).float().to(device)
    q_tok = torch.from_numpy(t_q_tok).long().to(device)
    q_num = torch.from_numpy(t["q_num"]).float().to(device)
    s_y   = torch.from_numpy(t["s_y"]).long().to(device)
    q_y   = torch.from_numpy(t["q_y"]).long().to(device)

    opt.zero_grad()

    s_z = encoder(s_tok, s_num)
    q_z = encoder(q_tok, q_num)

    logits_q, q_var = bayes_head(s_z, s_y, q_z)
    loss = F.cross_entropy(logits_q, q_y)
    loss.backward()
    nn.utils.clip_grad_norm_(encoder.parameters(), 5.0)
    nn.utils.clip_grad_norm_(bayes_head.parameters(), 5.0)
    opt.step()
    sched.step()

    train_losses.append(loss.item())

    with torch.no_grad():
        encoder.log_temp.data.clamp_(-3.0, 0.5)
        # Optional: clamp variances to reasonable range
        bayes_head.log_tau2.data.clamp_(-6.0, 4.0)
        bayes_head.log_sigma2.data.clamp_(-6.0, 4.0)

    if ep % EVAL_EVERY == 0:
        tr_acc, tr_std, _ = evaluate_bayes_metaoptnet(encoder, bayes_head, train_tasks[:50], device)
        va_acc, va_std, _ = evaluate_bayes_metaoptnet(encoder, bayes_head, val_tasks,       device)
        train_accs.append(tr_acc)
        val_accs.append(va_acc)
        mean_loss = float(np.mean(train_losses[-EVAL_EVERY:]))

        print(f"[{ep}/{EPISODES}] loss={mean_loss:.3f}  "
              f"train={tr_acc:.3f}±{tr_std:.3f}  val={va_acc:.3f}±{va_std:.3f}  "
              f"tau2={bayes_head.log_tau2.exp().item():.3e}  sigma2={bayes_head.log_sigma2.exp().item():.3e}")

        if va_acc > best_val:
            best_val, bad = va_acc, 0
            torch.save({
                "encoder": encoder.state_dict(),
                "head": bayes_head.state_dict()
            }, "best_bayes_metaoptnet_cnn.pt")
            print("  -> New best model saved!")
        else:
            bad += 1

        if bad >= PATIENCE:
            print("Early stopping.")
            break

print("Best Val Acc (Bayesian MetaOptNet-Ridge):", round(best_val, 4))

# ============================================================
# 11) Final test evaluation
# ============================================================
if os.path.exists("best_bayes_metaoptnet_cnn.pt"):
    ckpt = torch.load("best_bayes_metaoptnet_cnn.pt", map_location=device)
    encoder.load_state_dict(ckpt["encoder"])
    bayes_head.load_state_dict(ckpt["head"])

    test_acc, test_std, test_loss = evaluate_bayes_metaoptnet(encoder, bayes_head, test_tasks, device)
    print(f"TEST  acc={test_acc:.3f} ± {test_std:.3f} | loss={test_loss:.3f} | episodes={len(test_tasks)}")
else:
    print("No saved Bayesian MetaOptNet model found for testing")

print(f"Random baseline (1/N): {1.0/float(N):.3f}")

# ============================================================
# 12) Training plots (same style as before)
# ============================================================
fig, axes = plt.subplots(1, 2, figsize=(13, 5))

axes[0].plot(train_losses, alpha=0.3)
axes[0].plot(pd.Series(train_losses).rolling(window=20, min_periods=1).mean(), label="smoothed")
axes[0].set_title("Bayesian MetaOptNet-Ridge Training Loss")
axes[0].set_xlabel("Episode")
axes[0].legend()
axes[0].grid(True, alpha=0.3)

eval_steps = [i * EVAL_EVERY for i in range(1, len(val_accs) + 1)]
axes[1].plot(eval_steps, train_accs, marker='o', label="Train Acc")
axes[1].plot(eval_steps, val_accs, marker='s', label="Val Acc")
axes[1].axhline(1.0 / N, linestyle='--', label=f"Random {1.0/N:.3f}")
axes[1].axhline(best_val, linestyle=':', label=f"Best Val {best_val:.3f}")
axes[1].set_title("Bayesian MetaOptNet-Ridge Accuracies")
axes[1].set_xlabel("Episode")
axes[1].legend()
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# ============================
# Uncertainty-aware t-SNE for Bayesian MetaOptNet-Ridge
# ============================
from sklearn.manifold import TSNE
import seaborn as sns
import matplotlib.pyplot as plt

LABEL_COL = "group"          # we are at group-level
TSNE_PERPLEXITY = 30
RANDOM_STATE = 42

# ---- 1) Helper to get embeddings for any dataframe ----
def embed_df(df_sample, model, n_max=None):
    """
    Encode sequences in df_sample with the SAME pipeline as training
    and return (embeddings, labels, df_sample).

    If n_max is given, randomly subsample that many rows.
    """
    model.eval()
    if n_max is not None:
        df_sample = df_sample.sample(
            min(n_max, len(df_sample)),
            random_state=RANDOM_STATE
        ).reset_index(drop=True)
    else:
        df_sample = df_sample.reset_index(drop=True)

    # --- token ids ---
    ids = [encode_ids(s, KMER_K) for s in df_sample["sequence"].tolist()]
    tok = np.full((len(ids), MAX_LEN), stoi[PAD], dtype=np.int64)
    for i, seq in enumerate(ids):
        s = seq[:MAX_LEN]
        tok[i, :len(s)] = s

    # --- numeric features (use same normalization as training) ---
    nums = df_sample[["gc_content", "seq_len"]].to_numpy(np.float32)
    nums = (nums - num_mean) / num_std

    with torch.no_grad():
        Xt = torch.from_numpy(tok).long().to(device)
        Xn = torch.from_numpy(nums).float().to(device)
        feats = model(Xt, Xn).cpu().numpy()   # [N, d]

    labels = df_sample[LABEL_COL].values  # strings (group names)
    return feats, labels, df_sample


# ---- 2) Bayesian predictive variance in embedding space ----
def bayesian_predictive_variance(Zs, ys_int, Zq, tau2, sigma2):
    """
    Zs: [S, d] train embeddings (support)
    ys_int: [S] train labels as integer ids (0..C-1)
    Zq: [Q, d] test embeddings (queries)

    Returns:
        probs: [Q, C] softmax probabilities (from posterior mean)
        var_q: [Q]    scalar predictive variance per query
    """
    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    Zs_t = torch.from_numpy(Zs).float().to(device)
    ys_t = torch.from_numpy(ys_int).long().to(device)
    Zq_t = torch.from_numpy(Zq).float().to(device)

    S, d = Zs_t.shape
    num_classes = int(ys_t.max().item()) + 1

    # one-hot labels [S, C]
    Y = torch.nn.functional.one_hot(ys_t, num_classes=num_classes).float()

    # Bayesian ridge: A = Z^T Z / sigma2 + I / tau2
    #                 B = Z^T Y / sigma2
    tau2_t   = torch.tensor(tau2,   device=device, dtype=torch.float32)
    sigma2_t = torch.tensor(sigma2, device=device, dtype=torch.float32)

    A = (Zs_t.t() @ Zs_t) / sigma2_t + torch.eye(d, device=device) / tau2_t
    B = (Zs_t.t() @ Y)    / sigma2_t

    # posterior mean of weights: W = A^{-1} B
    W = torch.linalg.solve(A, B)          # [d, C]

    # posterior covariance of weights:
    # Σ_w = sigma2 * A^{-1}
    A_inv = torch.linalg.inv(A)
    Sigma_w = sigma2_t * A_inv            # [d, d]

    # predictive mean logits: Zq W
    logits = Zq_t @ W                     # [Q, C]
    probs  = torch.softmax(logits, dim=1)

    # predictive variance for each query (scalar per point):
    # var(y*) = z*^T Σ_w z* + sigma2
    tmp = Zq_t @ Sigma_w                  # [Q, d]
    var_q = (tmp * Zq_t).sum(dim=1) + sigma2_t  # [Q]

    return probs.cpu().numpy(), var_q.cpu().numpy()


# ---- 3) tau2, sigma2 from your training logs ----
TAU2   = 1.46e-01   # from your best-val epoch
SIGMA2 = 9.23e-01   # from your best-val epoch

# ---- 4) Get embeddings for subsets of train + test ----
Zs, ys, train_sub = embed_df(train_df, encoder, n_max=2000)   # ys are strings
Zq, yq, test_sub  = embed_df(test_df,  encoder, n_max=800)    # yq are strings

# Label-encode groups -> integer ids (shared mapping)
classes = np.unique(ys)
class_to_id = {c: i for i, c in enumerate(classes)}
ys_int = np.array([class_to_id[g] for g in ys], dtype=np.int64)
#yq_int = np.array([class_to_id[g] for g in yq], dtype=np.int64)  # not strictly needed here, but handy

# ---- 5) Bayesian predictive variance on test embeddings ----
probs_q, var_q = bayesian_predictive_variance(Zs, ys_int, Zq, tau2=TAU2, sigma2=SIGMA2)

# ---- 6) t-SNE on test embeddings ----
print(f"t-SNE (this may take a bit)... N = {len(yq)}, label={LABEL_COL}")
tsne = TSNE(
    n_components=2,
    perplexity=TSNE_PERPLEXITY,
    random_state=RANDOM_STATE,
    init="pca",
    learning_rate="auto",
)
reduced = tsne.fit_transform(Zq)

# ========================
# Plot A: colored by group (for reference)
# ========================
plt.figure(figsize=(10, 8))
sns.scatterplot(
    x=reduced[:, 0],
    y=reduced[:, 1],
    hue=yq,                      # still use string group names for the legend
    legend="full",
    palette="tab20",
    s=40,
    alpha=0.8,
)
plt.title("t-SNE of Token-CNN + Bayesian MetaOptNet embeddings (colored by group)")
plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left", ncol=1)
plt.tight_layout()
plt.show()

# ========================
# Plot B: colored by predictive variance (uncertainty)
# ========================
plt.figure(figsize=(10, 8))
sc = plt.scatter(
    reduced[:, 0],
    reduced[:, 1],
    c=var_q,
    s=40,
    cmap="viridis",
    alpha=0.9,
)
plt.colorbar(sc, label="Predictive variance (higher = more uncertain)")
plt.title("t-SNE of Bayesian MetaOptNet embeddings (colored by uncertainty)")
plt.xlabel("t-SNE 1")
plt.ylabel("t-SNE 2")
plt.tight_layout()
plt.show()
```
